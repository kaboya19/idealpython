```python
import os
import pandas as pd


klasor_yolu = "C:/iDeal/data"

dataframes = []

# Klasördeki tüm dosyaları listele ve txt dosyalarını seç
for dosya in os.listdir(klasor_yolu):
    if dosya.endswith('.txt'):  # Sadece .txt uzantılı dosyaları seç
        dosya_yolu = os.path.join(klasor_yolu, dosya)
        
        # Dosyayı oku (burada delimiter ve diğer parametreleri ayarlayabilirsiniz)
        df = pd.read_csv(
            dosya_yolu,
            delimiter=';',   # Noktalı virgül ile ayrıldığını varsayıyoruz
            decimal=',',     # Ondalık ayırıcı olarak virgül kullanılıyor
            header=None,     # Eğer sütun başlıkları yoksa
            names=['Tarih', 'Saat', 'Açılış', 'Yüksek', 'Düşük', 'Kapanış', 'Hacim']  # Sütun isimleri
        )
        
       
        df['Hisse'] = dosya[6:-7]
        
        # Listeye ekle
        dataframes.append(df)
        print(f"{dosya} verisi başarıyla okundu.")

# Tüm DataFrame'leri birleştir
tum_veri = pd.concat(dataframes, ignore_index=True)

# Tarih ve Saat sütunlarını birleştirip datetime formatına çevir, ardından indeks olarak ayarla
tum_veri['Datetime'] = pd.to_datetime(tum_veri['Tarih'] + ' ' + tum_veri['Saat'], format='%d.%m.%Y %H:%M')
tum_veri.set_index('Datetime', inplace=True)
tum_veri.drop(['Tarih', 'Saat'], axis=1, inplace=True)




tum_veri.to_csv("C:/iDeal/data/birlesik_veri.csv")
```
